```python
# Node Influence Power 
def NIP(i):
    cc = nx.clustering(G,i)
    deg = nx.degree(G,i)
    maxNode = i
    maxDeg = 0
    neighbors = nx.neighbors(G,i)
    for j in neighbors:
        if(nx.degree(G,j) > maxDeg):
            maxDeg=nx.degree(G,j)
            maxNode = j
    comm = set(nx.common_neighbors(G, i, maxNode))
    return deg - (len(comm) * (1-cc))

'''
rr = NIP(2) 
print(rr)
''' 
    
```

```python
def set_nodes_influence_power(G):
    nodes = G.nodes()
    for n in nodes:
        G.nodes[n]['nip']   = NIP(n)
 
```

```python
# Algorithm 1: Labeling nodes based on the Node Influence Power (NIP)
```

```python
def nodes_labeling(G):
    set_nodes_influence_power(G)
    # print(nx.get_node_attributes(G, "nip"))

    nx.set_node_attributes(G, 0, "flag")

    hasZeroFlag = True
    while hasZeroFlag:
        x = ''
        for n in G.nodes():
            if (G.nodes[n]['flag'] == 0):
                x = n
                break

        if(x == ''):
            break

        maxNeighborsNip = 0  
        y = ''
        xNeighbors = nx.neighbors(G,x)        
        for j in xNeighbors:
            if( G.nodes[x]['nip'] > G.nodes[j]['nip']):
                G.nodes[j]['flag'] = 1

            if( G.nodes[j]['nip'] > maxNeighborsNip):
                maxNeighborsNip = G.nodes[j]['nip']
                y = j

        if(G.nodes[x]['nip'] > G.nodes[y]['nip']):
            G.nodes[x]['flag'] = 2
            continue
        else:
            G.nodes[x]['nip'] = 1
            if (G.nodes[y]['flag']==2):
                continue 

        maxNeighborsNipY = 0  
        z = ''
        yNeighbors = nx.neighbors(G,y)        
        for j in yNeighbors:
            if( G.nodes[y]['nip'] > G.nodes[j]['nip']):
                G.nodes[j]['flag'] = 1  
            if( G.nodes[j]['nip'] > maxNeighborsNipY):
                maxNeighborsNipY = G.nodes[j]['nip']
                z = j    

        if(G.nodes[y]['nip'] > G.nodes[z]['nip']):
            G.nodes[y]['flag'] = 2  
            continue
        else:
            G.nodes[y]['flag'] = 1
            if (G.nodes[z]['flag']==2):
                continue

        G.nodes[z]['flag'] = 2
        zNeighbors = nx.neighbors(G,z)        
        for j in zNeighbors:
            if( G.nodes[z]['nip'] > G.nodes[j]['nip']):
                G.nodes[j]['flag'] = 1

        hasZeroFlag = any( v == 0 for k, v in nx.get_node_attributes(G, "flag").items())

'''
nodes_labeling(G)
print(nx.get_node_attributes(G, "flag"))
'''
nodes_labeling(G)
print(nx.get_node_attributes(G, "flag"))
    
```

```python
# Algorithm 2: Select the Candidate set
```

```python
# top k nodes for Influence Maximation
k=3  
def get_candidate_set(G,k): 
    candidateSet = []

    def filter_node2(n):
        return G.nodes[n]['flag'] == 2 
    G2 = nx.subgraph_view(G, filter_node=filter_node2)  

    for g in G2.nodes():
        candidateSet.append(g)  

    # print(candidateSet)
    #*****************************************************************************************    
    # Algorithm 2 : select "k" nodes from labeled nodes with "1" with highest Node Influence Power and lowest clustering coefficient
    #*****************************************************************************************


    tempSet = {}

    def filter_node1(n):
        return G.nodes[n]['flag'] ==1
    G1 = nx.subgraph_view(G, filter_node=filter_node1)


    # check to ensure is selected "k" is propertly valid or not.
    nodeCountG1 = nx.number_of_nodes(G1)
    if ( k > nodeCountG1 / 2 ): k = int(nodeCountG1 / 2 )

    sortG1 = sorted(G1.nodes(), key=lambda n: G1.nodes[n]['nip'], reverse=True)[:2*k]
    for i in sortG1:
        tempSet.update({i:nx.clustering(G,i)})

    candidateSet = candidateSet + sorted(tempSet, key=lambda n: tempSet[n], reverse=False)[:k]    

    FG = nx.Graph( nx.subgraph(G,candidateSet)) 
    # nx.draw(FG, pos, with_labels=True, node_size=900, font_size=18)

    #*****************************************************************************************    
    # Algorithm 2 : add some middle nodes to make just a one component for candidate set nodes.
    #*****************************************************************************************
    while (nx.is_connected(FG) == False):
        print(1)
        shortPath = []
        shortPathLength = nx.number_of_nodes(G)
        graphs = list(nx.connected_components(FG))
        for i in graphs[0]:
            for j in graphs[1]:
                shortPaths = nx.all_shortest_paths(G, source=i, target=j)
                for path in shortPaths:
                    sp = set(path) - set(graphs[0]).union(set(graphs[1]))
                    if (len(sp)<shortPathLength):
                        shortPath = path
                        shortPathLength = len(sp)

        nx.add_path(FG, shortPath) 
        
    return FG
 
FG = get_candidate_set(G,k)
```

```python
nx.draw(FG, pos, with_labels=True, node_size=900, font_size=18)
```

```python
### Algorithm 3 : Select the final seed nodes
```

```python
# def get_seed_nodes(G,k):

# nodes_labeling(G)
# get_candidate_set(G,k)

importances = {}  

for g in FG.nodes():
    nip = G.nodes[g]['nip']
    for n in nx.neighbors(FG,g):
        nip = nip + G.nodes[n]['nip']
    avg = nip/(G.degree(g)+1)
    sq = nx.square_clustering(G,g) 
    t = nx.subgraph(G,nx.neighbors(G,g)).number_of_edges() 
    imp = G.degree(g) - ( (math.exp(sq) * t) / avg )
    importances.update({g: imp})
    # print('{}: {}'.format(g,imp))

#print (importances) 
seedSet = sorted(importances, key=lambda n: importances[n], reverse=True)[:k] 
print(seedSet)

color_map = []
for node in G:
    if any( v == node for v in seedSet):
        color_map.append('green')
    else: 
        color_map.append('lightBlue') 
nx.draw(G, pos, node_color=color_map, with_labels=True, node_size=900, font_size=18)
        
```
